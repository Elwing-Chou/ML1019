```python
import pandas as pd
train = pd.read_csv("poem_train.csv", encoding="utf-8")
train
```

```python
test = pd.read_csv("poem_test.csv", encoding="utf-8")
test
```

```python
# 使用函式庫: 進去的東西一定要變成數字
# Series.replace(字典)
trans = {"李白":0, "杜甫":1, "白居易":2}
y_train = train["作者"].replace(trans)
y_test = test["作者"].replace(trans)
y_test
```

```python
# 重點: 分詞 -> 像是英文(我 喜歡 你)
import jieba
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。"
s = s.replace("\r", "").replace("\n", "")
" ".join(jieba.cut(s))
```

```python
# Pandas操作: apply
# Series.apply(自定義轉換流程)
def poem(s):
    s = s.replace("\r", "").replace("\n", "")
    return " ".join(jieba.cut(s))
# poem: 流程名字  poem(字串):()執行
# poem 不會直接執行, 到時候 apply 時候 poem(那格內容)
x_train = train["內容"].apply(poem)
x_test = test["內容"].apply(poem)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
```

```python
# 訓練: fit + transform
x_train_vec = vec.fit_transform(x_train)
# 測試資料: transform
x_test_vec = vec.transform(x_test)
```

```python
# items: (字典key, 字典value)
reverse = {v:k for k, v in vec.vocabulary_.items()}
reverse[29006]
```

```python
vec.vocabulary_
```

```python
print(x_train_vec)
```

```python
# 建立模型
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
pre = clf.predict(x_test_vec)
pre
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(pre, y_test)
```

```python
from sklearn.metrics import confusion_matrix
ori = ["李白", "杜甫", "白居易"]
c = [(author + "(預測)") for author in ori]
i = [(author + "(真實)") for author in ori]
mat = confusion_matrix(y_test, pre)
pd.DataFrame(mat, index=i, columns=c)
```
